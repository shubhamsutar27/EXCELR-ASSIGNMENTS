## FITTING A MODEL

```python
import statsmodels.formula.api as smf
model1 = smf.ols('Price~Age+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=toyota_cleaned3).fit()
```

```python
model1.params
```

```python
print(model1.tvalues, '\n', model1.pvalues)
```

For doors and cc , p value is greater than alpha.

## CALCULATING VIF

```python
rsq_Age = smf.ols('Age~KM+HP+Doors+Gears+Weight+cc+Quarterly_Tax',data=toyota_cleaned3).fit().rsquared
vif_Age = 1/(1-rsq_Age)

rsq_KM = smf.ols('KM~Age+HP+Doors+Gears+Weight+cc+Quarterly_Tax',data=toyota_cleaned3).fit().rsquared
vif_KM = 1/(1-rsq_KM)

rsq_HP = smf.ols('HP~Age+KM+Doors+Gears+Weight+cc+Quarterly_Tax',data=toyota_cleaned3).fit().rsquared
vif_HP = 1/(1-rsq_HP)

rsq_Doors = smf.ols('Doors~Age+KM+HP+Gears+Weight+cc+Quarterly_Tax',data=toyota_cleaned3).fit().rsquared
vif_Doors = 1/(1-rsq_Doors)

rsq_Gears = smf.ols('Gears~Age+KM+HP+Doors+Weight+cc+Quarterly_Tax',data=toyota_cleaned3).fit().rsquared
vif_Gears = 1/(1-rsq_Gears)

rsq_cc = smf.ols('cc~Age+KM+HP+Doors+Weight+Gears+Quarterly_Tax',data=toyota_cleaned3).fit().rsquared
vif_cc = 1/(1-rsq_cc)

rsq_Quarterly_Tax = smf.ols('Quarterly_Tax~Age+KM+HP+Doors+Weight+Gears+cc',data=toyota_cleaned3).fit().rsquared
vif_Quarterly_Tax = 1/(1-rsq_Quarterly_Tax)

rsq_Weight = smf.ols('Weight~Age+KM+HP+Doors+Gears+cc+Quarterly_Tax',data=toyota_cleaned3).fit().rsquared
vif_Weight = 1/(1-rsq_Weight)
```

```python
d1 = {'Variables':['Age','KM','HP','Doors','Gears','Weight','cc','Quarterly_Tax'],'VIF':[vif_Age,vif_KM,vif_HP,vif_Doors,vif_Gears,vif_Weight,vif_cc,vif_Quarterly_Tax]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

## MODEL DELETION DIAGNOSTICS

Detecting Influencers/Outliers

### COOK'S DISTANCE

```python
from statsmodels.graphics.regressionplots import influence_plot

model_influence = model1.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#plot the influencers values using stem plot
fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(toyota_cleaned3)),np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than 0.5
(np.argmax(c),np.max(c))
```

### HIGH INFLUENCE POINTS

```python
toyota_cleaned3.shape
```

```python
k = toyota_cleaned3.shape[1]
n = toyota_cleaned3.shape[0]
leverage_cutoff = 3*((k + 1)/n)
leverage_cutoff
```

```python
from statsmodels.graphics.regressionplots import influence_plot
import matplotlib.pyplot as plt

influence_plot(model1,alpha=0.5)

y = [i for i in range(-2,8)]
x = [leverage_cutoff for i in range(10)]
plt.plot(x,y,'r+')

plt.show()
```

From above plot, it is evident that data point 80 is influencer

```python
toyota_cleaned3[toyota_cleaned3.index.isin([80])]
```

## IMPROVING THE MODEL

```python
toyota1 = toyota_cleaned3
```

```python
#Discard the data point which are influencers and reassign the row number(reset_index)
toyota1=toyota1.drop(toyota1.index[80],axis=0).reset_index()
```

```python
toyota1
```

```python
#drop the original index
toyota1 = toyota1.drop(['index'],axis=1)
```

```python
fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(toyota1)),np.round(c,3));
plt.xlabel('Row index')
plt.ylabel('Cooks Distance');
plt.show()
```

```python
fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(toyota2)),np.round(c,3));
plt.xlabel('Row index')
plt.ylabel('Cooks Distance');
plt.show()
```

From above plot we can conclude that Cook's Distance is less than 1, so we can stop the process.
